```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Read the dataset into a pandas DataFrame
df = pd.read_csv('Marketing Campaign data.csv')

# Create a dictionary to store the metadata
metadata_dict = {
    'Attribute Name': df.columns.tolist(),
    'Data Type': df.dtypes.tolist(),
    'Description': [
        'Unique identifier for each customer record',
        'Gender of the customer (Male or Female)',
        'Age of the customer',
        'Marital status of the customer',
        'Country where the customer resides',
        'Income level of the customer',
        'Educational qualification of the customer',
        'Occupation of the customer',
        'Number of people in the customer`s household',
        'Number of years the customer has been living in the current residence',
        'Affinity Card status of the customer which also happens to be the target variable (1 = High-value, 0 = Low-value)',
        'Ownership of bulk pack diskettes (1 = Yes, 0 = No)',
        'Ownership of flat panel monitors (1 = Yes, 0 = No)',
        'Ownership of home theater packages (1 = Yes, 0 = No)',
        'Ownership of bookkeeping applications (1 = Yes, 0 = No)',
        'Ownership of printer supplies (1 = Yes, 0 = No)',
        'Ownership of Y box games (1 = Yes, 0 = No)',
        'Ownership of OS doc set kanji (1 = Yes, 0 = No)',
        'Comments of the customer on the shopping experience ']
}

# Create a pandas DataFrame from the metadata dictionary
metadata_df = pd.DataFrame(metadata_dict)

# Print the metadata table
print(metadata_df)
```

```python
#print unique values of each attribute 
print(df.CUST_ID.is_unique)

print(df.CUST_GENDER.unique())

print(df.AGE.unique())

print(df.COUNTRY_NAME.unique())

print(df.CUST_INCOME_LEVEL.unique())

print(df.EDUCATION.unique())

print(df.OCCUPATION.unique())

print(df.HOUSEHOLD_SIZE.unique())

print(df.YRS_RESIDENCE.unique())

print(df.AFFINITY_CARD.unique())

print(df.BULK_PACK_DISKETTES.unique())

print(df.FLAT_PANEL_MONITOR.unique())

print(df.HOME_THEATER_PACKAGE.unique())

print(df.BOOKKEEPING_APPLICATION.unique())

print(df.PRINTER_SUPPLIES.unique())

print(df.Y_BOX_GAMES.unique())

print(df.OS_DOC_SET_KANJI.unique())

#drop columns PRINTER_SUPPLIES and COMMENTS
df1=df.drop(['PRINTER_SUPPLIES','COMMENTS'],axis=1)
```

```python
# Calculate the correlation of each variable with the target variable
corr_matrix = df1.corr()
print(corr_matrix["AFFINITY_CARD"].sort_values(ascending=False))

#Drop columns having less corelation with the target variable
df2=df1.drop(['Y_BOX_GAMES','OS_DOC_SET_KANJI','FLAT_PANEL_MONITOR','BULK_PACK_DISKETTES'],axis=1)

#print the columns of the new dataframe to check whether only relevant columns exist
print(df2.columns)
```

```python
#find and print unique values of "OCCUPATION" with counts in descending order
occ_counts=df2.OCCUPATION.value_counts(ascending=False)
print(occ_counts)

#replace "?" by highest occuring value and print transformed values
df2['OCCUPATION']=df2['OCCUPATION'].replace(['?'],'Exec.')
occ_counts=df2.OCCUPATION.value_counts(ascending=False)
print(occ_counts)
```

```python
#conversion of "CUST_GENDER" to binary
from sklearn import preprocessing 
#make an instance of Label Encoder
label_encoder = preprocessing.LabelEncoder()
df2['CUST_GENDER'] = label_encoder.fit_transform(df2.CUST_GENDER)
print(df2['CUST_GENDER'])
```

```python
#conversion of "COUNTRY_NAME" to ordinal numbers
country_counts = df2['COUNTRY_NAME'].value_counts(ascending=False)
country_dict = {country: i+1 for i, country in enumerate(country_counts.index)}
df2['COUNTRY_NAME'] = df2['COUNTRY_NAME'].map(country_dict)
print(df2['COUNTRY_NAME'])
```

```python
#conversion of "CUST_INCOME_LEVEL" to ordinal numbers
income_level=df2['CUST_INCOME_LEVEL'].drop_duplicates().sort_values()
income_dict={income: i+1 for i, income in enumerate(income_level)}
df2['CUST_INCOME_LEVEL']=df2['CUST_INCOME_LEVEL'].map(income_dict)
print(df2['CUST_INCOME_LEVEL'])
```

```python
#conversion of "EDUCATION" to numbers based on descending order of occurrence
education=df2['EDUCATION'].drop_duplicates().sort_values()
education_levels_descending=['PhD','Profsc','Masters','Bach.','< Bach.','Assoc-V','Assoc-A','HS-grad','12th','11th','10th','9th','7th-8th','5th-6th','1st-4th','Presch.']
education_dict = {education: i+1 for i, education in enumerate(education_levels_descending)}
df2['EDUCATION'] = df2['EDUCATION'].map(education_dict)
print(df2['EDUCATION'])
```

```python
#conversion of"HOUSEHOLD_SIZE" to ordinal numbers
household_size=df2['HOUSEHOLD_SIZE'].drop_duplicates().sort_values()
house_size_categories=['1','2','3','4-5','6-8','9+']
size_dict = {size: i+1 for i, size in enumerate(house_size_categories)}
df2['HOUSEHOLD_SIZE']=df2['HOUSEHOLD_SIZE'].map(size_dict)
print(df2['HOUSEHOLD_SIZE'])
```

```python
#Calculate summary statistics for all variables
summary = df2.describe()

#Add skewness and kurtosis to the summary
summary.loc['skewness'] = df2.skew()
summary.loc['kurtosis'] = df2.kurtosis()

#Print the summary
print(summary)
```

```python
#conversion of"CUST_MARITAL_STATUS" to ordinal numbers
marital_status=df2['CUST_MARITAL_STATUS'].drop_duplicates()
status_dict={status: i+1 for i,status in enumerate(marital_status)}
df2['CUST_MARITAL_STATUS']=df2['CUST_MARITAL_STATUS'].map(status_dict)
print(df2['CUST_MARITAL_STATUS'])

#conversion of"OCCUPATION" to ordinal numbers
occupation=df2['OCCUPATION'].drop_duplicates()
occ_dict={occ: i+1 for i,occ in enumerate(occupation)}
df2['OCCUPATION']=df2['OCCUPATION'].map(occ_dict)
print(df2['OCCUPATION'])

#Calculate the correlation of each variable with the target variable
corr_matrix = df2.corr()
print(corr_matrix["AFFINITY_CARD"].sort_values(ascending=False))
```

```python
# Get a list of variable names
variables = list(df2.columns)

# Remove non-numeric variables
numeric_variables = []
for variable in variables:
    if pd.api.types.is_numeric_dtype(df2[variable]):
        numeric_variables.append(variable)

# Main loop
while True:
    # Ask the user to choose two variables
    print('Choose two variables to plot (type "exit" to quit):')
    for i, variable in enumerate(numeric_variables):
        print(f'{i+1}. {variable}')
    choice1 = input('First variable: ')
    if choice1 == 'exit':
        break
    choice2 = input('Second variable: ')
    if choice2 == 'exit':
        break
    
    # Check that the user's choices are valid
    try:
        index1 = int(choice1) - 1
        variable1 = numeric_variables[index1]
        index2 = int(choice2) - 1
        variable2 = numeric_variables[index2]
    except:
        print('Invalid choice. Please try again.')
        continue
    
    # Plot the scatter plot of the chosen variables
    plt.scatter(df2[variable1], df2[variable2])
    plt.title(f'Scatter plot of {variable1} vs {variable2}')
    plt.xlabel(variable1)
    plt.ylabel(variable2)
    plt.show()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# prepare the data
X = df2.drop(['AFFINITY_CARD'], axis=1)
y = df2['AFFINITY_CARD']

# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# build a logistic regression model
lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print('Accuracy score of logistic regression model:', accuracy_lr)

#Visualising the results
import statsmodels.api as sm
model = sm.Logit(y_train, X_train).fit()
print(model.summary())

#Find intercept and coefficient values
feature_names = df2.columns[:-1].values.tolist()
coefs = pd.DataFrame(
    lr_model.coef_.T,   
    index=feature_names,
    columns=["Coefficients"]
)

print(coefs)

#Visualising the Coefficients in a bar chart - Effects Plot
coefs.plot(kind="bar", figsize=(13, 6),title="Effects Plot")

```
